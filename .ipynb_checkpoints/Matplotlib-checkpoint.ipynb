```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,50,60]
```

```python
weigh_list = [32,62,70,75,65,96,85,64,63]
```

```python
plt.plot(age_list,weigh_list,"r")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Yaşa Göre Kilo")
plt.show()
#plt.savefig kayıt edebiliriz
```

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weigh_list)
np_age_list,np_weight_list
```

```python
plt.plot(np_age_list,np_weight_list,"b")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Yaşa Göre Kilo")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,20,40)
numpy_arr1
#eşit aralıkta artar
```

```python
numpy_arr2=numpy_arr1 ** 3
```

```python
plt.plot(numpy_arr1,numpy_arr2,"r*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1,numpy_arr2,"r")
plt.subplot(1,2,2)
plt.plot(numpy_arr2,numpy_arr1,"b")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3])
figure_axes.plot(numpy_arr1,numpy_arr2,"y")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
#arametre [0.1,0.1,0.3,0.3] şu anlama geliyor:
#0.1: soldan mesafe
#0.1: alttan mesafe
#0.3: genişlik
#0.3: yükseklik
```

```python
new_fig=plt.figure(dpi=100)
new_axes=new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arr1,numpy_arr1 ** 2,label="numpy array ** 2",linewidth=2.0)
new_axes.plot(numpy_arr1,numpy_arr1 ** 3,label="numpy array ** 3",linewidth=4.0)
new_axes.plot(numpy_arr1,numpy_arr1 +3,label="numpy array ** 3",linewidth=4.0)
new_axes.legend()
#loc vererek lokasyonunu belirleriz
plt.show()
```
