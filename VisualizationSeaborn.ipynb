```python
plt.scatter("Height","Weight",data=df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height Vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", data=df)   
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height", y="Weight", data=df)   
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight",hue="Sex", data=df)  
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
#cinsiyete gre dalm
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight",hue="Medal", data=df)  
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
#Madalyaya
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight",hue="Sex",style="Medal", data=df)  
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
#madalyaya gre gsteriyor
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight",hue="Sex",style="Medal",size="Age", data=df)  
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lineplot(x="Height",y="Weight",hue="Sex",data=df)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height",hue="Sex",data=df)
plt.ylabel("Frequency")
plt.title("Athletes Height Distributiom")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height",hue="Sex",data=df,kind="kde")
plt.ylabel("Frequency")
plt.title("Athletes Height Distributiom")
plt.show()
```

```python
sns.barplot(x="Medal",y="Height",hue="Sex",data=df)
plt.title("Medals by Height")
plt.show()
```
