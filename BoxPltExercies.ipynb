```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#box plot
#median,quartile(25%,50%,75%),min max 1.5 * IQR outlier
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])
plt.figure(figsize=(6,5)) # büyük veya küçük boyut yapıyor
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data value")
#plt.grid(True)
plt.show()
```

#CHATGPT YORUMU
#📊 Boxplot yorumu:

Medyan (ortadaki çizgi)

Ortanca değer 56 çıkıyor.

Bu yüzden kutunun ortasındaki çizgi 56’da.

Kutu (Q1–Q3 arası, yani ortadaki %50 veri)

Q1 ≈ 22–25 civarında

Q3 ≈ 56

Yani verilerin yarısı 25 ile 56 arasında yoğunlaşmış.

Whisker’lar (normal sınırdaki uç değerler)

Alt whisker yaklaşık 5’e kadar uzuyor.

Üst whisker yaklaşık 60’a kadar uzuyor.

Bu aralıkta olan değerler normal kabul ediliyor.

Outlier (aykırı değerler)

Senin bu veride bariz outlier yok çünkü whisker’lar 5–60 aralığını kapsıyor.

Yani tüm değerler normal dağılım sınırları içinde.

Veri dağılımı

Veri oldukça sağa çarpık (çoğu değer 56 etrafında toplanmış, ama küçük tarafta 5–20 gibi uzak değerler var).

Bu yüzden kutunun alt kısmı uzun, üst kısmı kısa görünüyor.

🔎 Genel yorum

Verilerin çoğu 56 civarında toplanmış.

Küçük değerler (5, 7, 9, 15, 20) veri setini sola doğru genişletmiş → bu da kutunun alt tarafını uzatmış.

Aykırı değer görünmüyor, yani dataset dengeli.

Eğer bu gerçek bir ölçüm verisi olsaydı, “56” değerinin baskın olduğunu (çok tekrarlandığını) ve verilerin alt tarafında daha fazla çeşitlilik olduğunu söyleyebilirdik.

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.figure(figsize=(13,5))

plt.subplot(1,2,1)
sns.boxplot(x="class",y="age",data=df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class",y="fare",data=df)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.tight_layout() #sıkışık durumlarda yapabilirsin
plt.show()
```
