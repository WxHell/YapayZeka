```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#box plot
#median,quartile(25%,50%,75%),min max 1.5 * IQR outlier
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])
plt.figure(figsize=(6,5)) # b√ºy√ºk veya k√º√ß√ºk boyut yapƒ±yor
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data value")
#plt.grid(True)
plt.show()
```

#CHATGPT YORUMU
#üìä Boxplot yorumu:

Medyan (ortadaki √ßizgi)

Ortanca deƒüer 56 √ßƒ±kƒ±yor.

Bu y√ºzden kutunun ortasƒ±ndaki √ßizgi 56‚Äôda.

Kutu (Q1‚ÄìQ3 arasƒ±, yani ortadaki %50 veri)

Q1 ‚âà 22‚Äì25 civarƒ±nda

Q3 ‚âà 56

Yani verilerin yarƒ±sƒ± 25 ile 56 arasƒ±nda yoƒüunla≈ümƒ±≈ü.

Whisker‚Äôlar (normal sƒ±nƒ±rdaki u√ß deƒüerler)

Alt whisker yakla≈üƒ±k 5‚Äôe kadar uzuyor.

√úst whisker yakla≈üƒ±k 60‚Äôa kadar uzuyor.

Bu aralƒ±kta olan deƒüerler normal kabul ediliyor.

Outlier (aykƒ±rƒ± deƒüerler)

Senin bu veride bariz outlier yok √ß√ºnk√º whisker‚Äôlar 5‚Äì60 aralƒ±ƒüƒ±nƒ± kapsƒ±yor.

Yani t√ºm deƒüerler normal daƒüƒ±lƒ±m sƒ±nƒ±rlarƒ± i√ßinde.

Veri daƒüƒ±lƒ±mƒ±

Veri olduk√ßa saƒüa √ßarpƒ±k (√ßoƒüu deƒüer 56 etrafƒ±nda toplanmƒ±≈ü, ama k√º√ß√ºk tarafta 5‚Äì20 gibi uzak deƒüerler var).

Bu y√ºzden kutunun alt kƒ±smƒ± uzun, √ºst kƒ±smƒ± kƒ±sa g√∂r√ºn√ºyor.

üîé Genel yorum

Verilerin √ßoƒüu 56 civarƒ±nda toplanmƒ±≈ü.

K√º√ß√ºk deƒüerler (5, 7, 9, 15, 20) veri setini sola doƒüru geni≈ületmi≈ü ‚Üí bu da kutunun alt tarafƒ±nƒ± uzatmƒ±≈ü.

Aykƒ±rƒ± deƒüer g√∂r√ºnm√ºyor, yani dataset dengeli.

Eƒüer bu ger√ßek bir √∂l√ß√ºm verisi olsaydƒ±, ‚Äú56‚Äù deƒüerinin baskƒ±n olduƒüunu (√ßok tekrarlandƒ±ƒüƒ±nƒ±) ve verilerin alt tarafƒ±nda daha fazla √ße≈üitlilik olduƒüunu s√∂yleyebilirdik.

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.figure(figsize=(13,5))

plt.subplot(1,2,1)
sns.boxplot(x="class",y="age",data=df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class",y="fare",data=df)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.tight_layout() #sƒ±kƒ±≈üƒ±k durumlarda yapabilirsin
plt.show()
```
