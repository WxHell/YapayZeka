```python
import numpy as np
import pandas as pd
```

```python
df =pd.read_csv("7-concat_data1.csv")
```

```python
df.head()
```

```python
df2=pd.read_csv("7-concat_data2.csv")
```

```python
df_concat=pd.concat([df,df2],ignore_index=True)
#Normalde (ignore_index=False, varsayılan):
#Her iki DataFrame’in mevcut indexleri korunur.
#Yani birleştirme sonrası indexler tekrarlı olabilir.

#ignore_index=True kullanıldığında:
#Mevcut indexler göz ardı edilir, yeni baştan 0’dan başlayarak ardışık indexler atanır.
```

```python
df_concat
```

```python
#merge
#merge, iki DataFrame’i ortak sütun(lar) veya index(ler) üzerinden birleştirmek için kullanılır.
#Yani SQL’deki JOIN işleminin Pandas’taki karşılığıdır.
```

```python
df_merge = pd.read_csv("7-merge_data1.csv")
```

```python
df_merge
```

```python
df_merge2=pd.read_csv("7-merge_data2.csv")
```

```python
df_merge2
```

```python
#merge - outer join
```

```python
df_merged_outer=pd.merge(df_merge,df_merge2,on="Employee_ID",how="outer")
```

```python
df_merged_outer
```

```python
#merge- left join
```

```python
df_merged_left=df_merged_outer=pd.merge(df_merge,df_merge2,on="Employee_ID",how="left")
#birinci tabloyu dataframe ana olarak kabul ikinci ona göre al eşleşmeyen varsa kabul etme demek
#left dediğimiz df_merge'tür
```

```python
df_merged_left
```

```python
df_merged_right=df_merged_outer=pd.merge(df_merge,df_merge2,on="Employee_ID",how="right")
```

```python
df_merged_right
```

```python
#merge - inner join
```

```python
df_merged=pd.merge(df_merge,df_merge2,on="Employee_ID",how="inner")
#her iki tablonun kesişimlerini alır
```

```python
df_merged
```
