```python
def salary_category(salary):
    if salary<50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["Salary_Category"]=df["Salary"].apply(salary_category)
```

```python
df["Performance_Score"].min()
```

```python
def performance(performance):
    if performance<2:
        return "Low"
    elif 2 <= performance < 3:
        return "Medium"
    else:
        return "High"
```

```python
df["Performance_Score"].apply(performance)
```

```python
df["Performance_Category"]=df["Performance_Score"].apply(performance)
```

```python
df
```

```python
def adjust_performance(experience):
    if experience>10:
        return 1
    else:
        return 0
```

```python
df["Experience"].apply(adjust_performance)
```

```python
df["Adjusted"]=df["Experience"].apply(adjust_performance)
```

```python
df["New_Score"]=df["Performance_Score"] + df["Adjusted"]
```

```python
df
```

```python
df["Formatted_Name"]=df["Name"].apply(lambda x : x.replace("_"," "))
```
